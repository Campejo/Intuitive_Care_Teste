```python
import pdfplumber
import pandas as pd
import zipfile
import os
```

Variáveis

```python
pdf_path = r'assets\anexo1.pdf'

csv_path = 'tabela_extraida.csv'

tabela_completa = []
```

Extraindo a tabela do PDF

```python
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        tables = page.extract_table()

        if tables:
            tabela_completa.extend(tables)
```

Criando um DataFrame para a tabela extraída do PDF

```python
try:
    df = pd.DataFrame(tabela_completa)
    print('Tabela extraída!')

except Exception as e:
    print('Nenhuma tabela encontrada no PDF')
```

Antes dos valores modificados

```python
df
```

Alterando valores do DataFrame

```python
df = df.replace('OD', 'Seg. Odontológica')
df = df.replace('AMB', 'Seg. Ambulatorial')
print('Valores da tabela  foram modificados')
```

Após valores modificados

```python
df
```

Convertendo para CSV

```python
try:
    df.to_csv(csv_path, index=False)
    print(f'Tabela salva em {csv_path}')

except Exception as e:
    print(f'Não foi possível passar para CSV: {csv_path}')
```

Zipando o arquivo

```python
zip_filename = 'Teste_Thiago_Campello.zip'

with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipf.write(csv_path, os.path.basename(csv_path))
    print(f'Adicionado ao ZIP: {csv_path}')
```
