```python
import pdfplumber
import pandas as pd
import zipfile
import os
import shutil
```

Variáveis

```python
pdf_path = r'assets\pdfs\anexo1.pdf'

csv_path = r'assets\arquivos_csv\tabela_extraida.csv'

tabela_completa = []
```

Extraindo a tabela do PDF

```python
try:
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            tables = page.extract_table()

            if tables:
                tabela_completa.extend(tables)

except Exception as e:
    print(f'Arquivo não encontrado. Erro: {e}')
```

Criando um DataFrame para a tabela extraída do PDF

```python
try:
    df = pd.DataFrame(tabela_completa)
    df.columns = df.iloc[0]
    df = df[1:]
    print('Tabela extraída!')

except Exception as e:
    print('Nenhuma tabela encontrada no PDF')
```

Antes dos valores modificados

```python
df
```

Alterando valores do DataFrame

```python
# df = df.replace('OD', 'Seg. Odontológica')
# df = df.replace('AMB', 'Seg. Ambulatorial')
df = df.replace({'OB': 'Seg. Odontológica', 'AMB': 'Seg. Ambulatorial'})
print('Valores da tabela  foram modificados')
```

Após valores modificados

```python
df
```

Convertendo para CSV

```python
try:
    df.to_csv(csv_path, index=False)
    print(f'Tabela salva em {csv_path}')

except Exception as e:
    print(f'Não foi possível passar para CSV: {csv_path}')
```

Zipando o arquivo

```python
zip_filename = 'Teste_Thiago_Campello.zip'
try:
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        zipf.write(csv_path, os.path.basename(csv_path))
        print(f'Adicionado ao ZIP: {csv_path}')

except Exception as e:
    print('Não foi possível localizar o arquivo.')
    print(e)
```

Movendo o arquivo ZIP para a pasta arquivos_zip

```python
origin_path = 'Teste_Thiago_Campello.zip'
destination_path = r'assets\arquivos_zip\Teste_Thiago_Campello.zip'

shutil.move(origin_path, destination_path)
```
